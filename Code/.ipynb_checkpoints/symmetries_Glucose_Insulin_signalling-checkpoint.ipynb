# Structural identifiability of a glucose model using Lie symmetries

We consider the following model of glucose signalling:
\begin{equation}
\begin{split}
\dot{q_{1}}&=u-p_{1}q_{1}-p_{2}g_{2}\,,\\
\dot{q_{2}}&=p_{3}q_{1}+p_{4}q_{2}\,.\\
\end{split}
\label{eq:glucose_system}
\end{equation}
The two states are glucose $q_{1}(t)$ and insulin $q_{2}(t)$ dependening on time $t$, and the input $u(t)$ corresponds the glucose entering from the digestive system. Also, we have four rate parameters $\theta=\left(k_{1},k_{2},k_{3},k_{4}\right)$, where we observe the following output
\begin{equation}
   y=\frac{q_{1}}{V_{p}}\,, 
\label{eq:output_glucose}
\end{equation}
which means that we observe a fraction of the glucose in the blood. In [1] they find some Lie symmetries of this sytem using an ansatz based approach. In this work, we will try to find these Lie symmetries without resorting to any ansatze. Since this is a system of two first order ODE, this implies that the symmetry group is infinite. To this end, we re-write this system as a single second order ODE. The key step for doing this is to note that we can solve the second ODE for $q_{1}$ in the following way:
\begin{equation}
q_{1}=\dot{q_{2}}-p_{3}q_{2}\,,
\label{eq:q_1}
\end{equation}
and by differentiating the second ODE, substituting the first ODE and then substituting this last equation gives us the following single second order ODE:
\begin{equation}
\ddot{q_{2}}=\phi(q_{2},\dot{q_{2}},\theta)=(p_{3}-p_{1}p_{4})\dot{q_{2}}+p_{4}(p_{1}p_{3}-p_{2})q_{2}+p_{4}u\,.
\label{eq:glucose_system_second}
\end{equation}
We are interested in the symmetries of this second order ODE as the symmetry group is finite. Specifically, we are interested in a symmetry:
\begin{equation}
\Gamma_{\epsilon}:(t,q_{2},p_{1},p_{2},p_{3},p_{4})\mapsto(\hat{t}(\epsilon),\hat{q_{2}}(\epsilon),\hat{p_{1}}(\epsilon),\hat{p_{2}}(\epsilon),\hat{p_{3}}(\epsilon),\hat{p_{4}}(\epsilon))\,,
\label{eq:glucose_symmetry}
\end{equation}
that is generated by the following vector field
\begin{equation}
X=\xi(t,q_{2},\theta)\partial_{t}+\eta(t,q_{2},\theta)\partial_{q_{2}}+\sum_{i=1}^{4}\chi_{i}(t,q_{2},\theta)\partial_{p_{i}}
\label{eq:X_glucose}
\end{equation}
The second prolongation of this vector field is given by:
\begin{equation}
X=X+\eta^{(1)}(t,q_{2},\dot{q_{2}},\theta)\partial_{\dot{q_{2}}}+\eta^{(2)}(t,q_{2},\dot{q_{2}},\ddot{q_{2}},\theta)\partial_{\ddot{q_{2}}}\,,
\label{eq:X_glucose_2}
\end{equation}
where the prolonged infinitesimals are given by [2]:
\begin{align}
    \eta^{(1)}(t,q_{2},\dot{q_{2}},\theta)=&\eta_{t}+(\eta_{q_{2}}-\xi_{t})\dot{q_{2}}-\xi_{q_{2}}\left(\dot{q_{2}}\right)^2,\label{eq:eta_1}\\
    \eta^{(2)}(t,q_{2},\dot{q_{2}},\ddot{q_{2}},\theta)=&\eta_{zz}+(2\eta_{zu}-\xi_{zz})u_{z}+\left(\eta_{uu}-2\xi_{zu}\right)\left(u_{z}\right)^2-\xi_{uu}\left(u_{z}\right)^3\nonumber\\
    &+\left\{\eta_u - 2\xi_z -3\xi_u u_{z}\right\}u_{zz}.\label{eq:eta_2}  
    \end{align}
Then, the linearised symmetry condition is given by:
\begin{equation}
\eta^{(2)}=(p_{3}-p_{1}p_{4})\eta^{(1)}+p_{4}(p_{1}p_{3}-p_{2})\eta+\sum_{i=1}^{4}\left(\frac{\partial\phi_{i}}{\partial p_{i}}\right)\chi_{i}=0\,\mathrm{whenever}\quad\ddot{q_{2}}=\phi(q_{2},\dot{q_{2}},\theta)=(p_{3}-p_{1}p_{4})\dot{q_{2}}+p_{4}(p_{1}p_{3}-p_{2})q_{2}+p_{4}u\,.
\label{eq:lin_sym_glucose}
\end{equation}
This will decompose into a set of determining equations stemming from the different powers of $\dot{q}$ which gives us a system of equations that we can hopefully solve. Now, it's quite time consuming to do all these tasks by hand, so let's try to let SymPy do it instead!

[2] Hydon, Peter E. Symmetry methods for differential equations: a beginner's guide. No. 22. Cambridge University Press, 2000.
